# FEA Copilot Hand-Calculation Walkthrough

This notebook mirrors the `hand_calc_comparison` example and recomputes the cantilever beam result used by the repository's `mock` mode.

## Problem setup

- Beam length: `1.0 m`
- Section: `0.1 m x 0.1 m`
- Material: steel with `E = 200 GPa`
- Load: `150 N` downward at the tip
- Boundary condition: cantilever

```python
F = 150.0
L = 1.0
b = 0.1
h = 0.1
E = 200e9

I = b * h**3 / 12.0
delta_max = F * L**3 / (3.0 * E * I)
sigma_max = F * L * (h / 2.0) / I

print({'I_m4': I, 'max_deflection_m': delta_max, 'max_stress_pa': sigma_max})
```

## Expected result

The values should match `expected_metrics.json` in this directory and the CLI output from `./run.sh`:

- `max_deflection_m = 3.0e-05`
- `max_stress_pa = 9.0e05`
